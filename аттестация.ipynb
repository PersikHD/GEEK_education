```python
sns.countplot(x='waterfront', data=kc)
plt.title('Распределение домов по наличию виды на набережную')
plt.xlabel('Вид на набережную')
plt.ylabel('Количество домов')
plt.xticks(ticks=[0, 1], labels=['Нет', 'Да'])
```

```python
sns.countplot(x='floors', data=kc)
plt.title('Распределение этажей домов')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
```

```python
sns.countplot(x='grade', data=kc)
plt.title('Распределение состояния домов')
plt.xlabel('Оценка качества')
plt.ylabel('Количество домов')
```
